```python
import great_expectations as ge
import pandas as pd
from pathlib import Path
from core import config, utils
```

```python
features_file_path = Path(config.DATA_DIR, config.FILE_NAME)
features_df = ge.read_csv(features_file_path)
features_df
```

```python
#Expectation Suite
features_df.expect_column_values_to_not_be_null(column="Amount")
features_df.expect_column_values_to_be_of_type(column="Amount", type_="float")
expectation_suite = features_df.get_expectation_suite()
```

```python
# Validate reference data
features_df.validate(expectation_suite=expectation_suite, only_return_failures=True)["statistics"]
```
